```python
# Import libraries here
import numpy as np
from sklearn.linear_model import LogisticRegression
import pandas as pd
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier

# [Example] Read datasets
wine_train = pd.read_csv("C:/Users/Aditya/Documents/Manipal Sem 2/Machine Learning/Do Select Assignment/Wine Tasting Dataset/wine_train.csv")
wine_test = pd.read_csv("C:/Users/Aditya/Documents/Manipal Sem 2/Machine Learning/Do Select Assignment/Wine Tasting Dataset/wine_test.csv")

conditions = [
    (wine_train['quality'] == 6),
    (wine_train['quality'] < 6),
    (wine_train['quality'] > 6 )]
choices = ['normal', 'bad', 'good']

wine_train['taste'] = np.select(conditions, choices)

out = wine_train['taste']
x= wine_train.drop(['quality','taste'],axis=1)

xval= wine_test.drop(['id','quality'],axis=1)
```

```python
from sklearn.ensemble import RandomForestClassifier

model_rf = RandomForestClassifier(random_state=100,n_estimators=300)
model_rf.fit(x, out)

pred_test = model_rf.predict(xval)

pred_test
```

```python
lreg= LogisticRegression(C=100,penalty='l2')

lreg.fit(x,out)


print(xval.shape)
out_predict=lreg.predict(xval)

print(out_predict)
```

```python
#creating prediction.csv
df = pd.DataFrame(columns=['id','taste'])
df['taste']=out_predict
df['id']=wine_test['id']

print(df.head())

df.to_csv("wine_prediction.csv")
```
