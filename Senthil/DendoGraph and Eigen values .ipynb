```python
#Dendo Graph
from scipy.cluster.hierarchy import linkage,dendrogram
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import normalize
```

```python
import pandas as pd
import numpy as np
```

```python
stocks=pd.read_csv('company-stock-movements-2010-2015-incl.csv', index_col=0)
nstocks=normalize(stocks)

companies=list(stocks.index)
merging=linkage(nstocks, method='complete')
plt.figure(figsize=(15,8))
dendrogram(merging, labels=companies)

plt.show()
```

```python
from numpy.linalg import eig
```

```python
A=np.array([[1,2],[3,4],[5,6]])
print(A)
```

```python
M=np.mean(A.T,axis=1)
```

```python
print(M)
```

```python
c=A-M
```

```python
print(c)
```

```python
v=np.cov(c.T)
```

```python
print(v)
```

```python
values,vectors=eig(v)
```

```python
print(vectors)
print(values)
```

```python
p =vectors.T.dot(c.T)
print(p.T)
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(2)
```

```python
pca.fit(A)
```

```python
print(pca.components_)
print(pca.explained_variance_)
```

```python
B=pca.transform(A)
```

```python
print(B)
```
