```python
x,y,z = im.shape
```

```python
from sklearn import cluster
import numpy as np
```

```python
img=im.reshape(x*y,z)

clust=cluster.KMeans(n_clusters=3)
clust.fit(img)
cent=clust.cluster_centers_
lab=clust.labels_

plt.figure()
plt.imshow(cent[lab].astype(int).reshape(x,y,z))

```

```python
final=cent[lab].reshape(x,y,z)

np.place(final,final==cent[1],0)
np.place(final,final==cent[2],0)
plt.figure()
plt.imshow(final.astype(int))
```

```python
final1=cent[lab].reshape(x,y,z)

np.place(final1,final1==cent[0],0)
np.place(final1,final1==cent[1],0)
plt.figure()
plt.imshow(final1.astype(int))
```

```python
final2=cent[lab].reshape(x,y,z)

np.place(final2,final2==cent[0],0)
np.place(final2,final2==cent[2],0)
plt.figure(figsize=(20,10))
plt.imshow(final2.astype(int))
```

```python
final=final.reshape(x*y*z,1)
final1=final1.reshape(x*y*z,1)
final2=final2.reshape(x*y*z,1)
print(final.shape)
print(final1.shape)
print(final2.shape)
```

```python
c=0
c1=0
c2=0
```

```python
for i in final:
    if i != 0:
        c=c+1
        
for j in final1:
    if j != 0:
        c1=c1+1
        
for k in final2:
    if k != 0:
        c2=c2+1
        
```

```python
pc= (c/384000)*100 
pc1= (c1/384000)*100 
pc2= (c2/384000)*100

print(pc,pc1,pc2,)
```
