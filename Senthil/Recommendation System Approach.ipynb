```python
movies=movies.fillna(0)
```

```python
print(movies.isnull().sum(axis=0))
```

```python
min = movies['num_user_for_reviews'].quantile(0.99)
```

```python
min
```

```python
movies[movies['num_user_for_reviews']>min].shape[0]
```

```python
top_1percent=movies[movies['num_user_for_reviews']>min]
```

```python
top_1percent=top_1percent.reset_index()
```

```python
top_1percent=top_1percent.drop(columns=['index'])
```

```python
top_1percent
```

```python
top_1percent['popularity_index']=top_1percent["num_user_for_reviews"]*top_1percent["imdb_score"]
```

```python
movies['popularity_index']=movies["num_user_for_reviews"]*movies["imdb_score"]
```

```python
top_1percent['popularity_index']
```

```python
top5_popularityindex=top_1percent[['movie_title','popularity_index','imdb_score']].sort_values(by='popularity_index',ascending=False).head(5)
```
